### Given the following dataset, containing information about a year's worth of weather. 
- Using this data, calculate the percent of time it was raining each month. 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("weather_2012.csv")
```

```python
data
```

```python
month_list = []
```

```python
for index, row in data.iterrows(): 
    date_i = row["Date/Time"]
    month = date_i.split("-")[1]
    month_list.append(month)
```

```python
data['Month'] = month_list
data
```

```python
data_by_month = data.groupby('Month')
```

```python
len(data_by_month)
```

```python
rain_percentage = {}
number_days_rain = []
for i, k in data_by_month: 
    print(i, k)
    
    for index, row in k.iterrows(): 
        weather = row["Weather"]
        if "Rain" in weather:
            number_days_rain.append("Rain")
    rain_percentage[i]= (len(number_days_rain) / len(k)) * 100
    number_days_rain = []
        
```
